```python
import sys
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
class Window(QMainWindow):
    def __init__(self):
        super(Window, self).__init__()
        self.resize(400,300)
        self.move(300,300)
        self.setWindowTitle("重写事件处理函数")
        self.btn = QPushButton("",self)
        self.btn.move(100,100)
        self.btn.resize(200,50)

    def mousePressEvent(self,event):
        x = event.x()
        y = event.y()
        text = f"x:{x},y:{y}"
        self.btn.setText(text)
    
    def keyReleaseEvent(self,event):
        """
        Qt.Key_Left、Qt.Key_Right、Qt.Key_Up、Qt.Key_Down
        分别是Qt定义的左、右、上、下键的枚举值。设置每次移动的距离为20
        """
        rect = self.btn.geometry()
        if event.key() == Qt.Key_Left:
            self.btn.setGeometry(rect.left()-20,rect.top(),
                              rect.width(),rect.height())
        elif event.key() == Qt.Key_Right:
            self.btn.setGeometry(rect.left()+20,rect.top(),
                              rect.width(),rect.height())
        elif event.key() == Qt.Key_Up:
            self.btn.setGeometry(rect.left(),rect.top()-20,
                              rect.width(),rect.height())
        elif event.key() == Qt.Key_Down:
            self.btn.setGeometry(rect.left(),rect.top()+20,
                              rect.width(),rect.height())

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = Window()
window.show()
app.exec()
```

```python
import sys
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication,QMainWindow,QPushButton
class Window(QMainWindow):
    """
    自定义信号必须在类创建时定义，不能在类创建后作为类的属性动态添加
    （如在_init_方法中添加）。
    """
    my_signal = pyqtSignal(object) 
    def __init__(self):
        super(Window, self).__init__()
        self.resize(400,300)
        self.move(300,300)
        self.setWindowTitle("信号与槽")
        self.btn = QPushButton("",self)
        self.btn.move(100,100)
        self.btn.resize(200,50)
        self.my_signal.connect(self.show_text)
        self.my_signal.connect(self.move_button)
    
    def mousePressEvent(self,event):
        self.my_signal.emit(event)
    
    def show_text(self,event):
        x = event.x()
        y = event.y()
        text = f"x:{x},y:{y}"
        self.btn.setText(text)
    
    def move_button(self,event):
        rect = self.btn.geometry()
        x = event.x()
        y = event.y()
        self.btn.setGeometry(x,y,rect.width(),rect.height())

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = Window()
window.show()
app.exec()
```

```python
import sys
from PyQt5 import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
class Window(QMainWindow):
    def __init__(self):
        super(Window, self).__init__()
        self.setWindowTitle("Image Viewer")
        self.resize(600, 400)
        btn1 = QPushButton("放大",self)
        btn1.move(20,20)
        btn2 = QPushButton("缩小",self)
        btn2.move(140,20)
        btn3 = QPushButton("全图",self)
        btn3.move(260,20)
        self.imageLabel = QLabel(self)
        self.imageLabel.resize(400,300)
        self.imageLabel.move(20,80)
        image = QImage(r"c:\data\photos\p1.jpg")
        #对原始图像按比例进行缩放，返回在标签控件中全图显示的图像
        self.fit_image = image.scaled(self.imageLabel.size(),1)
        pixmap = QPixmap.fromImage(self.fit_image)
        self.imageLabel.setPixmap(pixmap)
        btn1.clicked.connect(self.zoomIn)
        btn2.clicked.connect(self.zoomOut)
        btn3.clicked.connect(self.fit)

    def zoomIn(self):
        #计算当前图像与全图的比例，乘上1.2，再对全图图像进行放大
        image = self.imageLabel.pixmap().toImage()  
        scale = image.size().width()/self.fit_image.size().width()
        size = self.fit_image.size()*scale*1.2
        new_image = self.fit_image.scaled(size,1,1)
        self.imageLabel.setPixmap(QPixmap.fromImage(new_image))
    
    def zoomOut(self):
        #计算当前图像与全图的比例，乘上0.8，再对全图图像进行缩小
        image = self.imageLabel.pixmap().toImage() 
        scale = image.size().width()/self.fit_image.size().width()
        size = self.fit_image.size()*scale*0.8
        new_image = self.fit_image.scaled(size,1,1)
        self.imageLabel.setPixmap(QPixmap.fromImage(new_image))
    
    def fit(self):
        #显示全图图像
        pixmap = QPixmap.fromImage(self.fit_image)
        self.imageLabel.setPixmap(pixmap)

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = Window()
window.show()
app.exec()
```

```python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.QtGui import *
class Window(QMainWindow):
    def __init__(self):
        super(Window, self).__init__()
        self.resize(800,600)
        self.view1 = QGraphicsView(self)
        self.setCentralWidget(self.view1)
        self.scene = QGraphicsScene()
        rect1 = QRectF(100,100,200,200)
        self.item1 = QGraphicsRectItem(rect1)
        self.item1.setBrush(QBrush(QColor("Green"))) 
        self.scene.addItem(self.item1)
        self.radio_btn1 = QRadioButton("显示坐标",self)
        self.radio_btn1.setChecked(True)
        self.radio_btn1.move(20,20)
        self.radio_btn1.clicked.connect(self.show_coordinate)
        self.radio_btn2 = QRadioButton("移动图元",self)
        self.radio_btn2.move(150,20)
        self.radio_btn2.clicked.connect(self.move_item)
        self.btn1 = QPushButton("放大",self)
        self.btn1.move(280,20)
        self.btn1.clicked.connect(self.zoomIn)
        self.btn2 = QPushButton("缩小",self)
        self.btn2.move(410,20)
        self.btn2.clicked.connect(self.zoomOut)
        self.view1.setScene(self.scene)
        self.statusBar = self.statusBar()
    
    def show_coordinate(self,event):
        #使鼠标点击图元时，图元不被选中和移动
        self.item1.setFlag(QGraphicsItem.ItemIsSelectable,False)
        self.item1.setFlag(QGraphicsItem.ItemIsMovable,False)
    
    def move_item(self,event):
        #使鼠标点击图元时，图元能被选中和移动,同时清空状态栏中的信息
        self.item1.setFlag(QGraphicsItem.ItemIsSelectable,True)
        self.item1.setFlag(QGraphicsItem.ItemIsMovable,True)
        self.statusBar.showMessage("")
    
    def zoomIn(self,event):
        self.view1.scale(1.2, 1.2)

    def zoomOut(self,event):
        self.view1.scale(0.8, 0.8)
    
    def mousePressEvent(self,event):                        
        view_x = event.x()
        view_y = event.y()
        #视图坐标转场景坐标
        scene_x = self.view1.mapToScene(view_x,view_y).x()
        scene_y = self.view1.mapToScene(view_x,view_y).y()
        #场景坐标转图元坐标
        item_x = self.item1.mapFromScene(scene_x,scene_y).x()
        item_y = self.item1.mapFromScene(scene_x,scene_y).y()
        message = f"图元坐标：{item_x:.2f},{item_y:.2f}；\
        场景坐标：{scene_x:.2f},{scene_y:.2f}；\
        视图坐标：{view_x:.2f},{view_y:.2f}"
        self.statusBar.showMessage(message) 

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = Window()
window.show()
app.exec()
```

```python
import sys
import fiona
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
class Window(QMainWindow):
    def __init__(self):
        super(Window, self).__init__()
        self.resize(800,600)
        self.view = QGraphicsView(self)
        self.setCentralWidget(self.view)
        self.view.scene=QGraphicsScene()
        self.view.setScene(self.view.scene)
        #加载STATES、USHIGH、CAPITALS数据
        self.load_USA_STATES()
        self.load_USA_USHIGH()
        self.load_USA_CAPITALS()
        rect = self.view.sceneRect()  #返回场景范围矩形
        x_scale = self.width()/rect.width()
        y_scale = self.height()/rect.height()
        if x_scale <= y_scale:
            scale = x_scale
        else:
            scale = y_scale
        #使输入数据全图显示（留5%空隙）
        self.view.scale(scale*0.95,scale*0.95)
        
    def load_USA_STATES(self):
        collection = fiona.open(r'c:\data\USA\STATES.shp', 'r')
        for record in collection:
            name = record["properties"]["STATE_NAME"] #用于后一示例的属性查询
            coordinates = record['geometry']['coordinates']
            if record['geometry']['type'] == "Polygon":
                #增加一层嵌套
                coordinates = [coordinates]
            for part in coordinates:
                points = []
                for ring in part:               
                    for p in ring:                        
                        point =QPointF(p[0],-p[1])
                        points.append(point)
                    polygon = QPolygonF(points)   
                    item = QGraphicsPolygonItem(polygon)
                    item = QGraphicsPolygonItem(polygon)
                    #用于后一示例的属性查询
                    item.setData(1,f"图层：STATES，州名：{name}") 
                    pen = QPen()
                    pen.setColor(QColor("gray"))
                    pen.setWidthF(0.3)
                    item.setPen(pen)
                    brush = QBrush(QColor("LightGray"))
                    item.setBrush(brush)
                    self.view.scene.addItem(item)
    
    def load_USA_USHIGH(self):
        collection = fiona.open(r'c:\data\USA\USHIGH.shp', 'r')
        for record in collection:
            name = record["properties"]["ROUTE"]  #用于后面示例的属性查询
            coordinates = record['geometry']['coordinates']
            if record['geometry']['type'] == "LineString":
                coordinates = [coordinates]  #增加一层嵌套
            for part in coordinates:
                p_count = len(part)
                #当前点和下一个点构成一个线item，并添加到场景
                for i in range(p_count-1):
                    line =  QLineF(part[i][0],-part[i][1],part[i+1][0],-part[i+1][1])
                    item = QGraphicsLineItem(line)
                    #用于后面示例的属性查询
                    item.setData(1,f"图层：USHIGH，公路名：{name}") 
                    pen = QPen()
                    pen.setColor(QColor("brown"))
                    pen.setWidthF(0.3)
                    item.setPen(pen)
                    self.view.scene.addItem(item)

    def load_USA_CAPITALS(self):
        collection = fiona.open(r"c:\data\USA\CAPITALS.shp", 'r')
        #计算用于表示点的圆图元直径（每个像素对应的直径*像素个数）
        bound_width = collection.bounds[2] - collection.bounds[0]
        pixel_size = bound_width/self.width()
        pixel_number = 8   
        size = pixel_size*pixel_number
        for record in collection:
            name = record["properties"]["CITY_NAME"]  #用于后一示例的属性查询
            coordinates = record['geometry']['coordinates']
            x = coordinates[0]
            y = coordinates[1]
            topLeft_x = float(x) - size/2
            topLeft_y = -float(y) - size/2
            rect = QRectF(topLeft_x, topLeft_y,size,size)
            item = QGraphicsEllipseItem(rect)
            #用于后一示例的属性查询                       
            item.setData(1,f"图层：CAPITAL，州府名：{name}") 
            pen = QPen()
            pen.setWidth(0)
            item.setPen(pen) 
            brush = QBrush(QColor("red"))
            item.setBrush(brush) 
            self.view.scene.addItem(item)

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = Window()
window.show()
app.exec()
```

```python
class Window(QMainWindow): 
    #其它已有代码   
    def wheelEvent(self, event):
        if event.angleDelta().y() > 0:
            self.zoomIn()
        else:
            self.zoomOut()
    
    def zoomIn(self):
            delta = 0.1   #定义每次滚动缩放系数的变化量
            self.view.scale(1 + delta,1 + delta)

    def zoomOut(self):
            delta = 0.1   #定义每次滚动缩放系数的变化量
            self.view.scale(1 - delta,1 - delta)
```

### 利用PyQt开发桌面GIS软件

```python
import sys
from os import path
import fiona
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.QtGui import *
      
class GraphicsView(QGraphicsView):
    def __init__(self):
        super().__init__()
        self.rubberBand = QRubberBand(QRubberBand.Rectangle, self)
        self.setMouseTracking(True)    
    
    def mousePressEvent(self,event):
        self.setMouseTracking(False)
        self.origin_pos = event.pos()
        self.rubberBand.setGeometry(QRect(self.origin_pos, QSize())) 
        if window.action in ["zoomIn","zoomOut"]:
            self.rubberBand.show()
                
    def mouseMoveEvent(self,event):
        #显示鼠标位置的坐标
        if self.hasMouseTracking() == True:
            view_x = event.x()
            view_y = event.y()
            #视图坐标转场景坐标
            scene_x = round(window.view.mapToScene(view_x,view_y).x(),4)
            scene_y = round(window.view.mapToScene(view_x,view_y).y(),4)
            message = f"坐标：{scene_x},{-scene_y}"
            window.statusBar.showMessage(message)
        #改变方框大小
        elif window.action in ["zoomIn","zoomOut"]:
            #利用normalized()方法可以反方向拉框
            rect = QRect(self.origin_pos, event.pos()).normalized()
            self.rubberBand.setGeometry(rect)  
        
    def mouseReleaseEvent(self, event):
        self.setMouseTracking(True)
        self.end_pos = event.pos()
        rect = self.rubberBand.geometry()
        self.rubberBand.hide()
        
        if window.action in ["zoomIn","zoomOut"]:
            #先把视图的中心点平移到方框的中心点，再进行缩放
            width = rect.width()
            height = rect.height()
            center_x = rect.center().x()
            center_y = rect.center().y()
            x = self.mapToScene(center_x,center_y).x()
            y = self.mapToScene(center_x,center_y).y()
            self.centerOn(x, y)

            x_scale = self.width()/width
            y_scale = self.height()/height
            if x_scale <= y_scale:
                scale = x_scale
            else:
                scale = y_scale
            if window.action == "zoomIn":
                self.scale(scale,scale)
            elif window.action == "zoomOut":
                self.scale(1/scale,1/scale)

        elif window.action == "pan":
            end_pos = window.view.mapToScene(self.end_pos)
            origin_pos = window.view.mapToScene(self.origin_pos)
            dx = origin_pos.x() - end_pos.x()
            dy = origin_pos.y() - end_pos.y()
            rectF = window.view.sceneRect()
            rectF.translate(dx, dy)
            self.setSceneRect(rectF)
        
        elif window.action == "identify":
            x = event.x()
            y = event.y()
            point = window.view.mapToScene(x,y)
            item = window.view.scene.itemAt(point,window.view.transform())
            if item != None:
                data2 = item.data(2)  #item.data(2)记录了字段名和字段值
                headerLabels = ["字段","值"]
                table = Table("Identify",data2,headerLabels)
                hLayout = QHBoxLayout()
                hLayout.addWidget(table)
                window.setLayout(hLayout)          
                table.show()
                      
class TreeWidget(QTreeWidget):
    def __init__(self):
        super().__init__()
        self.setMinimumWidth(250)
        self.setHeaderHidden(True)  #隐藏头部
        
    def onItemClicked(self,layerItem,index):
        #获得TOC中点击图层的名称
        layerItemName = layerItem.text(0)
        #对场景中的所有图元进行循环，如是点击图层中的图元，
        #则根据选中状态切换设置可见或不可见
        items = window.view.scene.items()
        for item in items:
            if item.data(1) == layerItemName:
                if layerItem.checkState(0) == Qt.Checked:
                    item.setVisible(True)
                else:
                    item.setVisible(False)

class Table(QTableWidget):
    def __init__(self,title,attribute,headerLabels):
        super().__init__()
        self.setWindowTitle(title)
        self.resize(400,300)
        row_count = len(attribute)
        self.setRowCount(row_count)
        column_count = len(attribute[0])
        self.setColumnCount(column_count)
        self.setHorizontalHeaderLabels(headerLabels)
        self.verticalHeader().setVisible(False)
        i = 0
        for field in attribute:
            newItem = QTableWidgetItem(field[0])
            self.setItem(i,0,newItem)
            newItem = QTableWidgetItem(str(field[1]))
            self.setItem(i,1,newItem)
            i = i + 1
    
class MainWindow(QMainWindow):
    action = ""    
    def __init__(self):
        super().__init__()
        self.resize(900, 600)
        self.statusBar = QStatusBar()
        self.setStatusBar(self.statusBar)
        self.toolBar = QToolBar()
        self.addToolBar(Qt.TopToolBarArea, self.toolBar)
        self.addLayerAction = QAction()
        self.zoomInAction = QAction()
        self.zoomInAction.setCheckable(True)
        self.zoomOutAction = QAction()
        self.zoomOutAction.setCheckable(True)
        self.panAction = QAction(self)
        self.panAction.setCheckable(True)
        self.identifyAction = QAction(self)
        self.identifyAction.setCheckable(True)
        self.fitAction = QAction(self)
        self.toolBar.addAction(self.addLayerAction)
        self.toolBar.addAction(self.zoomInAction)
        self.toolBar.addAction(self.zoomOutAction)
        self.toolBar.addAction(self.panAction)
        self.toolBar.addAction(self.fitAction)
        self.toolBar.addAction(self.identifyAction)

        self.addLayerAction.triggered.connect(self.addLayer)
        self.zoomInAction.triggered.connect(self.zoomIn)
        self.zoomOutAction.triggered.connect(self.zoomOut)
        self.panAction.triggered.connect(self.pan)
        self.identifyAction.triggered.connect(self.identify)
        self.fitAction.triggered.connect(self.fit)

        self.setWindowTitle("myGIS")
        self.toolBar.setWindowTitle("toolBar")
        self.addLayerAction.setText("添加图层")
        self.addLayerAction.setToolTip("添加图层")
        self.zoomInAction.setText("放大")
        self.zoomInAction.setToolTip("放大")
        self.zoomOutAction.setText("缩小")
        self.zoomOutAction.setToolTip("缩小")
        self.panAction.setText("平移")
        self.panAction.setToolTip("平移")
        self.identifyAction.setText("属性查询")
        self.identifyAction.setToolTip("属性查询")
        self.fitAction.setText("全图")
        self.fitAction.setToolTip("全图")
        
        #建立工具组，使工具组中的工具不能同时选中
        toolGroup = QActionGroup(self)
        toolGroup.addAction(self.zoomInAction)
        toolGroup.addAction(self.zoomOutAction)
        toolGroup.addAction(self.panAction)
        toolGroup.addAction(self.identifyAction)
        
        #实例化GraphicsView
        self.view = GraphicsView()
        self.setCentralWidget(self.view)
        self.view.scene = QGraphicsScene()
        self.view.setScene(self.view.scene)
        
        #创建一个DockWidget，放置TreeView
        self.dock = QDockWidget()
        self.dock.setWindowTitle("Table Of Contents")
        self.treeWidget = TreeWidget()
        
        #点击图层项的复选框，连接onItemClicked槽函数
        self.treeWidget.itemClicked.connect(self.treeWidget.onItemClicked)
        self.dock.setWidget(self.treeWidget)
        #将DockWidget加到主窗口中，默认停靠在右边
        self.addDockWidget(Qt.LeftDockWidgetArea,self.dock)
    
    def addLayer(self):
        fname,_ = QFileDialog.getOpenFileName(self,"选择shapefile文件","","*.shp")
        if fname != None:
            basename = path.basename(fname)
            mainFileName = basename.split(".")[0]
            #新建一个treeWidgetItem，添加到treeWidget
            treeItem = QTreeWidgetItem()
            treeItem.setText(0,mainFileName)
            treeItem.setCheckState(0,Qt.Checked)
            self.treeWidget.addTopLevelItem(treeItem)

            collection = fiona.open(fname, 'r')
            left = collection.bounds[0]
            right = collection.bounds[2]
            bottom = collection.bounds[1]
            top = collection.bounds[3]
            width = right - left
            height = top - bottom
            self.rect = QRectF(left,-top,width,height) #场景范围框
            self.view.fitInView(self.rect,Qt.KeepAspectRatio)
             
            geometry = collection.schema['geometry']
            if geometry == "Point":
                self.addPoint(mainFileName,collection)
            elif geometry == "LineString":
                self.addPolyline(mainFileName,collection)
            elif geometry == "Polygon":
                self.addPolygon(mainFileName,collection)
     
    def addPoint(self,mainFileName,collection):
        #计算图元大小（每个像素的大小乘上像素数）
        width = self.rect.width()
        pixel_size = width/self.width()  
        pixel_number = 8   
        size = pixel_size*pixel_number   #
        for record in collection:
            #读取记录中每个字段名及字段值，并添加到attribute列表中
            properties = record["properties"]
            attribute = []
            for field in properties:
                attribute.append([field,properties[field]])
                
            coordinates = record['geometry']['coordinates']
            x = coordinates[0]
            y = coordinates[1]
            topLeft_x = float(x) - size/2
            topLeft_y = -float(y) - size/2
            rect = QRectF(topLeft_x, topLeft_y,size,size)
            item = QGraphicsEllipseItem(rect)
            pen = QPen()
            pen.setWidth(0)
            item.setPen(pen) 
            brush = QBrush(QColor("red"))
            item.setBrush(brush)
            item.setData(1,mainFileName)
            item.setData(2,attribute)
            item.setZValue(30)
            self.view.scene.addItem(item)       
                    
    def addPolyline(self,mainFileName,collection):
        for record in collection:
            #读取记录中每个字段的名称及字段值，并添加到attribute列表中
            properties = record["properties"]
            attribute = []
            for field in properties:
                attribute.append([field,properties[field]])
                
            coordinates = record['geometry']['coordinates']
            if record['geometry']['type'] == "LineString":
                coordinates = [coordinates]
            for part in coordinates:
                pts = []
                for p in part:
                    pts.append([p[0],-p[1]])
            
            for i in range(len(pts)-1):
                line =  QLineF(pts[i][0],pts[i][1],pts[i+1][0],pts[i+1][1])
                item = QGraphicsLineItem(line)
                pen = QPen()
                pen.setColor(QColor("brown"))
                pen.setWidthF(0.3)
                item.setPen(pen)
                item.setData(1,mainFileName)
                item.setData(2,attribute)
                item.setZValue(20)
                self.view.scene.addItem(item)       
                
    def addPolygon(self,mainFileName,collection):
        for record in collection:
            #读取记录中每个字段名及字段值，并添加到attribute列表中
            properties = record["properties"]
            attribute = []
            for field in properties:
                attribute.append([field,properties[field]])
                
            coordinates = record['geometry']['coordinates']
            if record['geometry']['type'] == "Polygon":
                #增加一层嵌套
                coordinates = [coordinates]
            for part in coordinates:
                points = []
                for ring in part:               
                    for p in ring:                        
                        point =QPointF(p[0],-p[1])
                        points.append(point)
                    polygon = QPolygonF(points)   
                    item = QGraphicsPolygonItem(polygon)
                    pen = QPen()
                    pen.setColor(QColor("gray"))
                    pen.setWidthF(0.3)
                    item.setPen(pen)
                    #item.setPen(QPen(QColor("Yellow"), 0))
                    brush = QBrush(QColor("LightGray"))
                    item.setBrush(brush)
                    item.setData(1,mainFileName)
                    item.setData(2,attribute)
                    item.setZValue(10)
                    self.view.scene.addItem(item)
                    
    def zoomIn(self):
        self.action = "zoomIn"
      
    def zoomOut(self):
        self.action = "zoomOut"
    
    def pan(self):
        self.action = "pan"
       
    def fit(self):
        rect = self.view.scene.itemsBoundingRect()
        self.view.fitInView(rect,Qt.KeepAspectRatio)
                
    def identify(self):
        self.action = "identify"

if not QApplication.instance():
    app = QApplication(sys.argv)
else:
    app = QApplication.instance()
window = MainWindow()
window.show()
app.exec()
```
